```python
import pandas as pd 
```

```python
data_frame = pd.read_csv(r'C:\Users\19452\Downloads\covid19_africa.csv')
data_frame
```

```python
import matplotlib.pyplot as plt
```

```python
data_frame['Region'].value_counts().plot.pie(autopct='%.1f%%',figsize=(4,4))
```

```python
data_frame["Region"].unique()
```

```python
data_frame_WA = data_frame[data_frame["Region"] == "Western Africa"]
data_frame_WA.head()
```

```python
X = data_frame_WA["Country"].unique()
X
```

```python
confirmed_Y = []
deaths_Y = []
recovered_Y = []
active_Y = []
for x in X:
    confirmed_Y.append(data_frame_WA[data_frame_WA["Country"] == x]["Confirmed"].sum())
    deaths_Y.append(data_frame_WA[data_frame_WA["Country"] == x]["Deaths"].sum())
    recovered_Y.append(data_frame_WA[data_frame_WA["Country"] == x]["Recovered"].sum())
    active_Y.append(data_frame_WA[data_frame_WA["Country"] == x]["Active"].sum())
```

```python
import numpy as np
```

```python
X_axis = np.arange(len(X))
plt.rcParams["figure.figsize"] = (20,10)
plt.bar(X_axis - 0.2, confirmed_Y, 0.4, label = 'Confirmed')
plt.bar(X_axis - 0.2, deaths_Y, 0.4, label = 'Deaths')
plt.bar(X_axis + 0.2, recovered_Y, 0.4, label = 'Recovered')
plt.bar(X_axis + 0.2, active_Y,0.4, label='Active')
plt.xticks(X_axis, X)
plt.title("Plot of Confirmed, Recovered, Deceased, Active Cases")
plt.legend()
plt.show()
```
